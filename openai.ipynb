```python
import requests

API_KEY = "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"
MODEL = "openrouter/auto" # Automatically picks the best/cheapest model

def ask_llm(question, messages=[]):
    messages.append({"role": "user", "content": question})
    
    response = requests.post(
        url="https://openrouter.ai/api/v1/chat/completions",
        headers={"Authorization": f"Bearer {API_KEY}"},
        json={
            "model": MODEL,
            "messages": messages[-10:] # Long memory: keeps last 10 exchanges
        }
    ).json()

    answer = response["choices"][0]["message"]["content"]
    messages.append({"role": "assistant", "content": answer})
    return answer
```
