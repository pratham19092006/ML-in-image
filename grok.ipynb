```python
import requests

API_KEY = "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"

url = "https://api.groq.com/openai/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}

# conversation memory
messages = []

def ask_llm(question):

    global messages

    # add user message
    messages.append({
        "role": "user",
        "content": question
    })

    data = {
        "model": "llama-3.1-8b-instant",

        "messages": messages,

        "max_tokens": 1000
    }

    response = requests.post(
        url,
        headers=headers,
        json=data
    )

    result = response.json()

    answer = result["choices"][0]["message"]["content"]

    # save assistant response
    messages.append({
        "role": "assistant",
        "content": answer
    })

    print("Successful")
    return answer
```
