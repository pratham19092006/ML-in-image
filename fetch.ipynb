```python
import requests
import json

url = "https://raw.githubusercontent.com/pratham19092006/ML-in-image/refs/heads/main/1.ipynb" # 1 or 2 or 3
content = requests.get(url).text

notebook = json.loads(content)

all_code = ""

for cell in notebook["cells"]:

    if cell["cell_type"] == "code":

        code = "".join(cell["source"])

        all_code += code + "\n\n"

print(all_code)
```
